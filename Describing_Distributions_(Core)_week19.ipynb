```python
#imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from scipy import stats
```

```python
df = pd.read_csv('data (4).csv')
df.head()
```

First Analyse VitD_levels

###For each feature listed:
1.Plot a histogram with a kde (kernel density estimate)
- Add a line for the mean (red)
- Add a line for the median (green)
- Add a line for for +1 std from the mean (black)
- Add a line for the - 1 std from the mean (black)
- Highlight the range between +1 and =1 std (yellow)

```python
col = 'VitD_levels'
```

```python
feature = df[col]
mean = feature.mean()
median = feature.median()

std = feature.std()
plus_one_std = mean + std
minus_one_std = mean - std

fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(feature ,ax=ax,stat='density', kde = True)
ax.axvline(mean,color = 'red', label=f'Mean = {mean:,.2f}',)#line for mean
ax.axvline(median, ls='--', color = 'green', label = f'Median = {median:,.2f}')#Line for median

ax.axvline(plus_one_std, color = 'black',label=f'+1 std = {plus_one_std:,.2f}') #line for +1 std
ax.axvline(minus_one_std, color = 'black', label = f'-1 std = {minus_one_std:,.2f}') #line for -1 std

#Highlight the range between +1 and =1 std (yellow)
ax.axvspan(plus_one_std, minus_one_std, color = 'yellow')
ax.set_title(f'{col}')
ax.legend();
```
